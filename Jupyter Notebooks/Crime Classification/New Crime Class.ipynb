```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import sklearn

crime_data = pd.read_csv('E:/University/Fall 2019/COMP 4480 - Directed Studies/Final Data/Partly Combined Data/Combined_data.csv')
```

```python
crime_data
```

```python
# Import Pre-processing module
from sklearn import preprocessing

# Create Label Encoder
le = preprocessing.LabelEncoder()

# Convert the string columns to numeric
# .astype(str) to ensude they are read as strings and not as mixed types
crime_data['vc_hundred_block'] = le.fit_transform(crime_data['vc_hundred_block'].astype(str))
crime_data['vc_neighbourhood'] = le.fit_transform(crime_data['vc_neighbourhood'].astype(str))
crime_data['vwh_weather'] = le.fit_transform(crime_data['vwh_weather'].astype(str))
crime_data['vc_type'] = le.fit_transform(crime_data['vc_type'])

# Display the results
crime_data
```

```python
crime_data.dtypes
```

```python
# Encoding temperature to be above 0 for chi squared test
crime_data['vwh_temp'] = le.fit_transform(crime_data['vwh_temp'])

crime_data
```

```python
# split for training and testing
from sklearn.model_selection import train_test_split

# Select columns to that are data
cols = [col for col in crime_data if col not in ['vc_type']]

X = crime_data[cols]

# Select target data
y = crime_data['vc_type']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=100, test_size=0.3)
print(X_train.shape)
```

```python
# Tutorial via https://towardsdatascience.com/feature-selection-using-python-for-classification-problem-b5f00a1c7028

# Check for zero variance features
from sklearn.feature_selection import VarianceThreshold

sel_variance_threshold = VarianceThreshold() 
X_train_remove_variance = sel_variance_threshold.fit_transform(X_train)
print(X_train_remove_variance.shape)
```

```python
# Use a chi squared test to test relationship between each feature and the target

from sklearn.feature_selection import SelectKBest, chi2

sel_chi2 = SelectKBest(chi2, k=4)    # select 4 features
X_train_chi2 = sel_chi2.fit_transform(X_train, y_train)
print(sel_chi2.get_support())
```

Chi2: 4 Best True Columns: vc_minute, vc_hundred_block, vwh_temp, vwh_weather

```python
# try f test
from sklearn.feature_selection import SelectKBest, f_classif

sel_f = SelectKBest(f_classif, k=4)
X_train_f = sel_f.fit_transform(X_train, y_train)
print(sel_f.get_support())
```

f test: 4 Best True Columns: vc_year, vc_minute, vc_hundred_block, vc_neighbourhood, 

```python
# mutual_info_classif test
from sklearn.feature_selection import mutual_info_classif

sel_mutual = SelectKBest(mutual_info_classif, k=4)
X_train_mutual = sel_mutual.fit_transform(X_train, y_train)
print(sel_mutual.get_support())
```

mutual_info_classif: 4 Best True Columns: vc_hour, vc_minute, vc_hundred_block, vc_neighbourhood

But all the above are when we have 8 target options.  Lets simplify it to 2

```python
# Encode all Break and Enter's as 1 and everything else as 2

crime_simple_type = pd.read_csv('E:/University/Fall 2019/COMP 4480 - Directed Studies/Final Data/Partly Combined Data/Combined_data.csv')

crime_simple_type = crime_simple_type.replace(['Mischief', 'Theft from Vehicle',
       'Vehicle Collision or Pedestrian Struck (with Injury)',
       'Theft of Vehicle', 'Theft of Bicycle', 'Other Theft',
       'Vehicle Collision or Pedestrian Struck (with Fatality)'], 0)

crime_simple_type = crime_simple_type.replace(['Break and Enter Residential/Other', 'Break and Enter Commercial'], 1)

# Import Pre-processing module
from sklearn import preprocessing

# Create Label Encoder
le = preprocessing.LabelEncoder()

# Convert the string columns to numeric
# .astype(str) to ensude they are read as strings and not as mixed types
crime_simple_type['vc_hundred_block'] = le.fit_transform(crime_simple_type['vc_hundred_block'].astype(str))
crime_simple_type['vc_neighbourhood'] = le.fit_transform(crime_simple_type['vc_neighbourhood'].astype(str))
crime_simple_type['vwh_weather'] = le.fit_transform(crime_simple_type['vwh_weather'].astype(str))

# Encoding temperature to be above 0 for chi squared test
crime_simple_type['vwh_temp'] = le.fit_transform(crime_simple_type['vwh_temp'])

crime_simple_type
```

Now, test again

```python
# split for training and testing
from sklearn.model_selection import train_test_split

# Select columns to that are data
cols = [col for col in crime_simple_type if col not in ['vc_type']]

X = crime_simple_type[cols]

# Select target data
y = crime_simple_type['vc_type']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=100, test_size=0.3)
print(X_train.shape)
```

```python
# Check for zero variance features
from sklearn.feature_selection import VarianceThreshold

sel_variance_threshold = VarianceThreshold() 
X_train_remove_variance = sel_variance_threshold.fit_transform(X_train)
print(X_train_remove_variance.shape)
```

```python
crime_simple_type['vwh_temp'].unique()

```

```python
# Use a chi squared test to test relationship between each feature and the target

from sklearn.feature_selection import SelectKBest, chi2

sel_chi2 = SelectKBest(chi2, k=4)    # select 4 features
X_train_chi2 = sel_chi2.fit_transform(X_train, y_train)
print(sel_chi2.get_support())
```

Chi2: 4 Best True Columns: vc_hour, vc_hundred_block, vwh_temp, vwh_weather

Same as 8 Target Values

```python
# try f test
from sklearn.feature_selection import SelectKBest, f_classif

sel_f = SelectKBest(f_classif, k=4)
X_train_f = sel_f.fit_transform(X_train, y_train)
print(sel_f.get_support())
```

Different from 8 options!

f test: 4 Best True Columns: vc_year, vc_hour, vc_hundred_block, vwh_temp

```python
# mutual_info_classif test
from sklearn.feature_selection import mutual_info_classif

sel_mutual = SelectKBest(mutual_info_classif, k=4)
X_train_mutual = sel_mutual.fit_transform(X_train, y_train)
print(sel_mutual.get_support())
```

Different from 8 options!

mutual_info_classif: 4 Best True Columns: vc_hour, vc_hundred_block, vc_neighbourhood, vwh_visibility

```python
# Trying Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

model_tree = RandomForestClassifier(random_state=100, n_estimators=50)
model_tree.fit(X_train, y_train)
print(model_tree.feature_importances_)
sel_model_tree = SelectFromModel(estimator=model_tree, prefit=True, threshold='mean')  
      # since we already fit the data, we specify prefit option here
      # Features whose importance is greater or equal to the threshold are kept while the others are discarded.
X_train_sfm_tree = sel_model_tree.transform(X_train)
print(sel_model_tree.get_support())

```
