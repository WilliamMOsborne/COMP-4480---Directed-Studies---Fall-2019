```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import sklearn

random_blocks = pd.read_csv('E:/Vancouver Crime Final/Synthetic Data/random_with_spread.csv')

random_blocks
```

```python
blocks = random_blocks.values.tolist()
```

```python
len(blocks)
```

```python
blocks
```

```python
import random
import utm

random_blocks = []


for block in blocks:
    new_block = []
    new_block.append(block[0])
    new_block.append(block[1])
    new_block.append(block[2] + random.uniform(0, block[3]))
    new_block.append(block[4] + random.uniform(0, block[5]))
    
    latlong = utm.to_latlon(float(new_block[2]),float(new_block[3]), 10, 'U')
    new_block.append(latlong[0])
    new_block.append(latlong[1])    
    
    random_blocks.append(new_block)
    
len(random_blocks)
```

```python
random_blocks
```

```python
from pandas import DataFrame

random_df = DataFrame(random_blocks,columns=['r_id', 'hundred_block', 'utm_x', 'utm_y', 'latitude', 'longitude'])

random_df
```

```python
random_df.to_csv('E:/Vancouver Crime Final/Synthetic Data/random_blocks_with_location.csv')
```
